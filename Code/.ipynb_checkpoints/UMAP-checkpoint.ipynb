```python
import umap
import numpy as np
from mnist.loader import MNIST
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline
import seaborn as sns
sns.set(style='white', context='poster')
```

```python
print('Loading Fashion MNIST dataset')
(train, train_labels), (test, test_labels) = tf.keras.datasets.fashion_mnist.load_data()

data = np.array(np.vstack([train, test]), dtype=np.float64) / 255.0
data = np.reshape(data, (-1, 28*28))  # reshaping data for UMAP
target = np.hstack([train_labels, test_labels])
classes = [
    'T-shirt/top',
    'Trouser',
    'Pullover',
    'Dress',
    'Coat',
    'Sandal',
    'Shirt',
    'Sneaker',
    'Bag',
    'Ankle boot']
```

```python
%%time
embedding = umap.UMAP(n_neighbors=5).fit_transform(data)
```

```python
fig, ax = plt.subplots(1, figsize=(14, 10))
plt.scatter(*embedding.T, s=0.3, c=target, cmap='Spectral', alpha=1.0)
plt.setp(ax, xticks=[], yticks=[])
cbar = plt.colorbar(boundaries=np.arange(11)-0.5)
cbar.set_ticks(np.arange(10))
cbar.set_ticklabels(classes)
# plt.title('Fashion MNIST Embedded via UMAP');
plt.savefig("Fashion_MNIST_UMAP.pdf", format="pdf", bbox_inches="tight")
```
